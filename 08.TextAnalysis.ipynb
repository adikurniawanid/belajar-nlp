```python
import nltk
```

```python
def create_nltk_text(file):
    f = open(file,'r')
    raw = f.read()
    tokens = nltk.word_tokenize(raw)
    return nltk.Text(tokens)
```

```python
textA = create_nltk_text('a.txt')
textB = create_nltk_text('b.txt')
textC = create_nltk_text('c.txt')
```
